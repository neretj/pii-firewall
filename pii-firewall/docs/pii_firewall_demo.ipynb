# PII Firewall Demo

Try **pii-firewall** in your browser. No local install needed.

This notebook shows three real-world cases:
1. Healthcare chatbot
2. Fintech support ticket
3. Legal document review

Run each cell with `Shift + Enter`.

```
# Install with regex-only backend (no heavy ML models in this demo)
%pip install pii-firewall -q
```

```
from privacy_firewall import create_firewall

# Helper LLM function (in production, use OpenAI, Anthropic, etc.)
def mock_llm(prompt: str) -> str:
    return f"Understood: {prompt[:120]}..."

firewall = create_firewall("healthcare", detector_backend="regex")
```

## 1. Healthcare chatbot

```
patient_input = "Patient John Doe, SSN 123-45-6789, diagnosed with hypertension. Prescribed lisinopril 10mg."

result = firewall.secure_call(
    text=patient_input,
    context={
        "tenant_id": "hospital-001",
        "case_id": "patient-123",
        "thread_id": "consultation-1",
        "actor_id": "doctor-456",
    },
    llm_client=mock_llm
)

print("→ LLM saw:")
print(result.sanitized_text)
print("
→ User got:")
print(result.final_text)
```

## 2. Fintech support ticket

```
ticket = "My IBAN is ES91 2100 0418 4502 0005 1332 and my card ends in 4242. Email: john@example.com"

finance = create_firewall("finance", detector_backend="regex")

result = finance.secure_call(
    text=ticket,
    context={"tenant_id": "acme", "case_id": "t-987", "thread_id": "th-1", "actor_id": "agent-1"},
    llm_client=mock_llm
)

print("→ LLM saw:")
print(result.sanitized_text)
print("
→ User got:")
print(result.final_text)
```

## 3. Legal document review

```
clause = "On 2024-05-12, Acme Corp signed an agreement with John Doe at 100 Market St, San Francisco."

legal = create_firewall("legal", detector_backend="regex")

result = legal.secure_call(
    text=clause,
    context={"tenant_id": "law-firm-001", "case_id": "case-77", "thread_id": "doc-1", "actor_id": "paralegal-3"},
    llm_client=mock_llm
)

print("→ LLM saw:")
print(result.sanitized_text)
print("
→ User got:")
print(result.final_text)
```
